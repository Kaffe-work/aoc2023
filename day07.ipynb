```python
import math
from collections import Counter
with open("day07.txt") as f:
    lines = f.readlines()
```

```python
hands = tuple((x[0], int(x[1])) for x in map(
    lambda x: x.strip().split(), lines))
#print(hands)


def get_numerical(card):
    if card.isdigit():
        return int(card)
    return {'T': 10, 'J': 11, 'Q': 12, 'K': 13, 'A': 14}[card]
newhands = []
for hand in hands:
  replacement = []
  for card in hand[0]:
    replacement.append((get_numerical(card)))
  newhands.append((replacement, hand[1]))

hands2 = newhands
def get_strength(hand):
    trick_strength = tuple(sorted(Counter(hand).values(), reverse=True))
    return trick_strength, hand

newhands.sort(key=lambda p: get_strength(p[0]))
print(newhands)
result = sum(pair[1] * (i+1) for i, pair in enumerate(newhands))

trick_str = []
def get_joke_strength(hand):
    numberofjokers = hand.count(11)
    unjoker = tuple(card for card in hand if card != 11)
    trick_stra, junk = get_strength(unjoker)
    if not trick_stra:
      trick_str = (5,)
    else :

      trick_str = list(trick_stra)
      trick_str[0] += numberofjokers
    trick_str = tuple(trick_str)

    new_hand = tuple(1 if card == 11 else card for card in hand)
    return trick_str, new_hand

hands2.sort(key=lambda p: get_joke_strength(p[0]))
result2 = sum(pair[1] * (i+1) for i, pair in enumerate(hands2))
print(f"solution to 1 is: {result}")
print(f"solution to 1 is: {result2}")
```
